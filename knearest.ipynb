```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
sns.set(color_codes=True)
```

```python
#the deterministic function
def F(x):
    return np.sin(x**2)
```

```python
sigma = 1/10.
N = 10000
```

```python
#the noisy output value for a value x
#of the form Y = f(X) + eps, where eps~N(0,100)

def Y(x):
    return F(x) + sigma*np.random.randn()
```

```python
#generate some data of the form Y = f(X) + eps, where eps~N(0,100)


xs = 20*np.random.rand(N)-10



ys = F(xs) + sigma*np.random.randn(N)
```

```python
plt.plot(xs, ys, '.')
```

```python
def knearest(point, xs, k):
    #infinities = np.inf*np.ones(k)
    #zeros = np.zeros(k)
    #neighbours = zip(zeros, infinities)
    distances = (point - xs)**2
    idxs = np.argsort(distances)[0:k]
    
    #
    return xs[idxs]
```

```python
#do k-nearest neighbours 

k=1
testdata = 20*np.random.rand(N)-10
predictions = []
for d in testdata:
    # find k nearest
    #print d
    #print xs
    neighbours = knearest(d, xs, k)
    
    y = 0
    for n in neighbours:
        y += Y(n)
    y = y/k
    predictions.append([d,y])
preds = np.array(predictions)
plt.plot(preds[:,0], preds[:,1], '.')
```
