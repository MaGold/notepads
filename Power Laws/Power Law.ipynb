# Pareto

```python
a, m = 3., 2.  # shape and mode
s = (np.random.pareto(a, 1000000) + 1) * m
p = plt.hist(s, bins=100)
```

```python
inds = np.arange(10,50,0.001)
counts = np.zeros(inds.size)

c=0
for i in inds:
    counts[c] = (s[s>i]).size
    c = c+1
```

```python
p = plt.plot(inds, counts)
```

```python
logcounts = np.log10(counts)
logcounts[logcounts<0] = 0

p = plt.plot(np.log(inds), logcounts)
```

# Power law?

The probability density function for powerlaw is:

powerlaw.pdf(x, a) = a * x^(a-1)

for 0 <= x <= 1, a > 0.

```python
from scipy.stats import powerlaw
a = 1.0
r = powerlaw.rvs(a, size=100000)
r = a/r
p = plt.hist(r, bins=1000, range=(0,2))
```

```python
inds = np.arange(0.75,1.0,0.001)
counts = np.zeros(inds.size)

c=0
for i in inds:
    counts[c] = (r[r>i]).size
    c = c+1
    
p = plt.plot(inds, counts)
```

```python
logcounts = np.log10(counts)
logcounts[logcounts<0] = 0

p = plt.plot(np.log(inds), logcounts)
```

# Uniform to Power law?

```python
r = np.random.uniform(size=100000)

xmin = 0.01
alpha = -2.5
#x = xmin * (1-r) ** (-1/(alpha-1))
x = xmin * (r) ** (alpha)

p = plt.hist(x, bins=100, range=(0,2))
```

```python
inds = np.arange(0.0,50.0,0.01)
counts = np.zeros(inds.size)

c=0
for i in inds:
    counts[c] = (x[x>i]).size
    c = c+1
    
p = plt.plot(inds, counts)
```

```python
logcounts = np.log10(counts)
logcounts[logcounts<0] = 0

p = plt.plot(np.log(inds), logcounts)
```

# Cumulative distribituion

from Newman's paper

```python
r = np.random.uniform(size=1000000)

xmin = 1
alpha = 2.5
x = xmin * (1-r) ** (-1/(alpha-1))
#x = xmin * (r) ** (alpha)

p = plt.plot(x)
```

```python
bins = np.arange(1,8.1,0.1)
hist, bin_edges = np.histogram(x, bins=bins, density=False)
p = plt.plot(bin_edges[:-1],hist)
```

```python
bins = np.arange(1,20000,0.1)
hist, bin_edges = np.histogram(x, bins=bins, density=False)
p = plt.plot(np.log10(bin_edges[:-1]),np.log10(hist))
```
