##$t(\nu)$-distribution:
If $z \sim N(0,1)$ and $\chi \sim \chi^2(\nu)$ and they are independent, then $$t=\frac{z}{\left[\chi / \nu\right]^{1/2}}$$ follows the $t$-distribution.

```python
t = []
for i in range(len(chi)):
    t.append( np.random.randn() / np.sqrt(chi[i]/nu))
```

```python
p = plt.hist(t, bins=np.linspace(-5, 5, 100))
```
