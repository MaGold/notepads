```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
sns.set(color_codes=True)
```

```python
n, p = 100, .6  # number of trials, probability of each trial
N = 1000000
s_binom = np.random.binomial(n, p, N)
s_poisson = np.random.poisson(n*p, N)
```

```python
bins = np.arange(0, 2*n*p, 1)
hist, bin_edges = np.histogram(s_binom, bins=bins, density=False)
hist=np.array(hist)
fig, ax = plt.subplots(1,1)
plot=ax.plot(hist)
```

```python
bins = np.arange(0, 2*n*p, 1)
hist2, bin_edges = np.histogram(s_poisson, bins=bins, density=False)
hist2=np.array(hist2)
fig, ax = plt.subplots(1,1)
plot=ax.plot(hist, label='Bin(n,p)')
plot=ax.plot(hist2, label='Po(np)')

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)

plt.xlim([n*p/1.5, n*p + (n*p - n*p/1.5)])
```
