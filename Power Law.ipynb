```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
sns.set(color_codes=True)
```

# Simple Power function

```python
x = np.arange(1000)
y = x*x
p = plt.plot(x, y)
```

```python
p = plt.plot(np.log(x), np.log(y))
```

# Pareto

```python
a, m = 3., 2.  # shape and mode
s = (np.random.pareto(a, 1000000) + 1) * m
p = plt.hist(s, bins=100)
```

```python
inds = np.arange(10,50,0.001)
counts = np.zeros(inds.size)

c=0
for i in inds:
    counts[c] = (s[s>i]).size
    c = c+1
```

```python
p = plt.plot(inds, counts)
```

```python
logcounts = np.log10(counts)
logcounts[logcounts<0] = 0

p = plt.plot(np.log(inds), logcounts)
```

# Normal Distribution

```python
x = np.random.randn(1000000)
```

```python
p = plt.hist(x, bins=100)
```

```python
inds = np.arange(4,5,0.001)
counts = np.zeros(inds.size)

c=0
for i in inds:
    counts[c] = (x[x>i]).size
    c = c+1
    
p = plt.plot(inds, counts)
```

```python
logcounts = np.log10(counts)
logcounts[logcounts<0] = 0

p = plt.plot(np.log(inds), logcounts)
```

# Binomial

```python
x = np.random.binomial(n=1000, p=1/2.0, size=1000000)
p = plt.hist(x, bins=1000)
```

```python
inds = np.arange(500,600,1)
counts = np.zeros(inds.size)

c=0
for i in inds:
    counts[c] = (x[x>i]).size
    c = c+1
    
p = plt.plot(inds, counts)
```

```python
logcounts = np.log10(counts)
logcounts[logcounts<0] = 0

p = plt.plot(np.log(inds), logcounts)
```

# Poisson

```python
x = np.random.poisson(lam=50, size=1000000)
# print(x)
p = plt.hist(x, bins=1000, range=(0,100))
```

```python
inds = np.arange(60,80,1)
counts = np.zeros(inds.size)

c=0
for i in inds:
    counts[c] = (x[x>i]).size
    c = c+1
    
p = plt.plot(inds, counts)
```

```python
logcounts = np.log10(counts)
logcounts[logcounts<0] = 0

p = plt.plot(np.log(inds), logcounts)
```
