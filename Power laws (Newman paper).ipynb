Implementation of some of the content in https://arxiv.org/abs/cond-mat/0412004

```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
sns.set(color_codes=True)
```

```python
# setup ticks for plots

t1 = np.linspace(1, 9, 9)
t2 = t1*10
t3 = t2*10
t4 = t3*10
t5 = t4*10

t = np.concatenate((t1, t2, t3, t4))
```

# Generate data from a power law

```python
r = np.random.uniform(size=1000000)

xmin = 1
alpha = 2.5
x = xmin * (1-r) ** (-1/(alpha-1))

p = plt.plot(x)
```

```python
bins = np.arange(1,8.1,0.1)
hist, bin_edges = np.histogram(x, bins=bins, density=True) # density True makes area under curve equal to 1
p = plt.plot(bin_edges[:-1],hist)
```

# Fig 3 b) 

Log plot of the data

```python
bins = np.arange(1,20000,0.1)
hist, bin_edges = np.histogram(x, bins=bins, density=False)
hist=np.array(hist)
fig, ax = plt.subplots(1,1)
p=ax.plot(bin_edges[:-1], hist/float(hist[0]))

ax.set_yscale('log')
ax.set_xscale('log')
plt.xticks(t, rotation='vertical')
plt.xlim([1,150])
plt.ylim([1.0/10**5,1])
```

# Fig 3 c)

"logarithmic binning"

```python
bins = np.arange(1,50,1)
i=2
for b in bins[2:]:
    bins[i] = bins[i-1] + 2 * (bins[i-1] - bins[i-2])
    i = i+1
bins[:5]
```

```python
# log-scaled bins
bins = np.logspace(0, 4, 50)  # 50 samples from 10^0 to 10^3
widths = (bins[1:] - bins[:-1])

# Calculate histogram
hist, bin_edges = np.histogram(x, bins=bins)

# normalize by bin width
hist_norm = hist/widths

# plot it!
#plt.bar(bins[:-1], hist_norm, widths)

plt.plot(bins[:-1], hist_norm/sum(hist))

plt.xscale('log')
plt.yscale('log')
z=plt.xticks(t, rotation='vertical')

plt.ylim([1.0/10**9,1])
plt.xlim([1,2500])
```

```python
xsorted = np.sort(x)
```
